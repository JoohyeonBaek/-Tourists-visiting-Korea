```python
# 관광객 증가율

# print(tourist_country_2015)
# print("-------")
# print(tourist_country_2016)

increaseValue = []


def getIncrease(last, this):
    return (this-last)/last * 100
    
# 2015 - 2016
increaseCountry = []
increaseValue = []

increaseCountry = tourist_country_2015["국적"].values.tolist()
length = len(tourist_country_2015.values)
for i in range(0,length):
    increaseValue.append(getIncrease(tourist_country_2015.values[i][1], tourist_country_2016.values[i][1]))

increaseDict = sorted(dict(zip(increaseCountry, increaseValue)).items(),
                     reverse=True,
                     key=lambda item: item[1])

finalIncreaseCountry = []
finalIncreaseValue = []


for i in increaseDict:
    finalIncreaseCountry.append(i[0])
    finalIncreaseValue.append(i[1])
    
print(increaseDict)
    
plt.figure(figsize=(50,50))
x = np.arange(length)
plt.bar(x, finalIncreaseValue)
plt.xticks(x, finalIncreaseCountry)
plt.show()

# 2016-2017

increaseCountry = []
increaseValue = []

increaseCountry = tourist_country_2016["국적"].values.tolist()
length = len(tourist_country_2016.values)
for i in range(0,length):
    increaseValue.append(getIncrease(tourist_country_2016.values[i][1], tourist_country_2017.values[i][1]))

increaseDict = sorted(dict(zip(increaseCountry, increaseValue)).items(),
                     reverse=True,
                     key=lambda item: item[1])

print(increaseDict)

finalIncreaseCountry = []
finalIncreaseValue = []

for i in increaseDict:
    finalIncreaseCountry.append(i[0])
    finalIncreaseValue.append(i[1])

plt.figure(figsize=(50,50))
x = np.arange(length)
plt.bar(x, finalIncreaseValue)
plt.xticks(x, finalIncreaseCountry)
plt.show()

# 2017-2018

increaseCountry = []
increaseValue = []

increaseCountry = tourist_country_2017["국적"].values.tolist()
length = len(tourist_country_2017.values)
for i in range(0,length):
    increaseValue.append(getIncrease(tourist_country_2017.values[i][1], tourist_country_2018.values[i][1]))

increaseDict = sorted(dict(zip(increaseCountry, increaseValue)).items(),
                     reverse=True,
                     key=lambda item: item[1])

print(increaseDict)

finalIncreaseCountry = []
finalIncreaseValue = []

for i in increaseDict:
    finalIncreaseCountry.append(i[0])
    finalIncreaseValue.append(i[1])
    
plt.figure(figsize=(50,50))
x = np.arange(length)
plt.bar(x, finalIncreaseValue)
plt.xticks(x, finalIncreaseCountry)
plt.show()

# 2018-2019

increaseCountry = []
increaseValue = []

increaseCountry = tourist_country_2017["국적"].values.tolist()
length = len(tourist_country_2018.values)
for i in range(0,length):
    increaseValue.append(getIncrease(tourist_country_2018.values[i][1], tourist_country_2019.values[i][1]))

increaseDict = sorted(dict(zip(increaseCountry, increaseValue)).items(),
                     reverse=True,
                     key=lambda item: item[1])

print(increaseDict)

finalIncreaseCountry = []
finalIncreaseValue = []

for i in increaseDict:
    finalIncreaseCountry.append(i[0])
    finalIncreaseValue.append(i[1])
    
plt.figure(figsize=(50,50))
x = np.arange(length)
plt.bar(x, finalIncreaseValue)
plt.xticks(x, finalIncreaseCountry)
plt.show()
```
