```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
```

```python
thailand_all = pd.read_excel("./data/thailand/thailand_all.xlsx")
thailand_age = pd.read_csv("./data/thailand/thailand_age.csv", thousands = ',')
thailand_gender = pd.read_csv("./data/thailand/thailand_gender.csv")
thailand_visit = pd.read_csv("./data/thailand/thailand_visit.csv")
thailand_activity = pd.read_csv("./data/thailand/thailand_activity.csv")
```

```python
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.figure(figsize=(10,10))
x = np.arange(12)
plt.title("월별 한국 관광객 수 (태국)")
plt.bar(x, thailand_all["계"].values.tolist()[:12])
plt.xticks(x, thailand_all["월"].values.tolist()[:12])
plt.show()
```

```python
# print(philippine_age)

ageData = []

ageData.append(thailand_age["20세이하"].values.tolist()[0])
ageData.append(thailand_age["20대"].values.tolist()[0])
ageData.append(thailand_age["30대"].values.tolist()[0])
ageData.append(thailand_age["40대"].values.tolist()[0])
ageData.append(thailand_age["50대"].values.tolist()[0])
ageData.append(thailand_age["60대"].values.tolist()[0])
print(ageData)


plt.rc('axes', titlesize=20)
plt.figure(figsize=(8,8))
ratio = ageData
labels = ['20세 이하', '20대', '30대', '40대', '50대', '60대']

wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.title("방한 관광객 연령 비율 (필리핀)")
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,  wedgeprops=wedgeprops)
plt.show()
```

```python
plt.rc('axes', titlesize=20)
plt.figure(figsize=(8,8))
ratio = [167220, 338495]
labels = ['남성', '여성']

wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.title("방한 관광객 남여 비율 (태국)")
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,  wedgeprops=wedgeprops)
plt.show()
```

```python
plt.figure(figsize=(10,10))
length = len(thailand_visit)
x = np.arange(length)
plt.title("방한 관광객 방문 지역 (필리핀)")
plt.bar(x, thailand_visit["2019"].values.tolist())
plt.xticks(x, thailand_visit["방문지역"].values.tolist())
plt.show()
```
