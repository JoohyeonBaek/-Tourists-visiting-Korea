```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
```

```python
vietnam_all = pd.read_excel("./data/vietnam/vietnam_all.xlsx")
vietnam_age = pd.read_csv("./data/vietnam/vietnam_age.csv", thousands = ',')
vietnam_gender = pd.read_csv("./data/vietnam/vietnam_gender.csv")
vietnam_visit = pd.read_csv("./data/vietnam/vietnam_visit.csv")
vietnam_activity = pd.read_csv("./data/vietnam/vietnam_activity.csv")
```

```python
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.figure(figsize=(10,10))
x = np.arange(12)
plt.title("월별 한국 관광객 수 (베트남)")
plt.bar(x, vietnam_all["계"].values.tolist()[:12])
plt.xticks(x, vietnam_all["월"].values.tolist()[:12])
plt.show()
```

```python
# print(philippine_age)

ageData = []

ageData.append(vietnam_age["20세이하"].values.tolist()[0])
ageData.append(vietnam_age["20대"].values.tolist()[0])
ageData.append(vietnam_age["30대"].values.tolist()[0])
ageData.append(vietnam_age["40대"].values.tolist()[0])
ageData.append(vietnam_age["50대"].values.tolist()[0])
ageData.append(vietnam_age["60대"].values.tolist()[0])
print(ageData)


plt.rc('axes', titlesize=20)
plt.figure(figsize=(8,8))
ratio = ageData
labels = ['20세 이하', '20대', '30대', '40대', '50대', '60대']

wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.title("방한 관광객 연령 비율 (베트남)")
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,  wedgeprops=wedgeprops)
plt.show()
```

```python
plt.rc('axes', titlesize=20)
plt.figure(figsize=(8,8))
ratio = [230005, 262069]
labels = ['남성', '여성']

wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.title("방한 관광객 남여 비율 (베트남)")
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False,  wedgeprops=wedgeprops)
plt.show()
```

```python
plt.figure(figsize=(10,10))
length = len(vietnam_visit)
x = np.arange(length)
plt.title("방한 관광객 방문 지역 (필리핀)")
plt.bar(x, vietnam_visit["2019"].values.tolist())
plt.xticks(x, vietnam_visit["방문지역"].values.tolist())
plt.show()
```
